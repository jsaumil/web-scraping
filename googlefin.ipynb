```python
from dataclasses import dataclass
```

```python
@dataclass
class stock:
    ticker: str
    exchange: str
    price:float=0
    currency:str="INR"
    inr_price: float=0
    
    def __post_init__(self):
     price_info=get_price_information(self.ticker,self.exchange)
    
     if price_info["ticker"]==self.ticker:
         self.price=price_info["price"]
         self.currency=price_info["currency"]
         self.inr_price=price_info["inr_price"]
```

```python
@dataclass
class position:
    stock: stock
    quantity:int
```

```python
@dataclass
class portfolio:
    positions: list[position]
    
    def get_total_value(self):
        total_value=0
        
        for position in self.positions:
            total_value += position.quantity * position.stock.inr_price
            
        return total_value
```

```python
def get_fx_to_inr(currency):
    fx_url=f"https://www.google.com/finance/quote/{currency}-INR"
    resp=r.get(fx_url)
    soup=BeautifulSoup(resp.content,"lxml")
    
    fx_rate=soup.find("div",{"data-last-price":True})
    fx=float(fx_rate["data-last-price"])
    return fx
```

```python
def get_price_information(ticker,exchange):
    url=f"https://www.google.com/finance/quote/{ticker}:{exchange}"
    resp=r.get(url)
    soup=BeautifulSoup(resp.content,"lxml")
    
    price_div=soup.find("div",attrs={"data-last-price":True})
    price=float(price_div["data-last-price"])
    currency=price_div["data-currency-code"]
    
    
    inr_price=price
    if currency != "INR":
        fx=get_fx_to_inr(currency)
        inr_price=price*fx
    
    
    return{
    "ticker":ticker,
    "exchange":exchange,
    "price":price,
    "currency":currency,
    "inr_price":inr_price
}
```

```python
def display_portfolio_summary(p):
    if not isinstance(p,portfolio):
        raise TypeError("Please provide an instance of the portfolio type")
    
    portfolio_value=p.get_total_value()
    
    position_data=[]
    
    for position in p.positions:
        position_data.append([
            position.stock.ticker,
            position.stock.exchange,
            position.quantity,
            position.stock.exchange,
            position.quantity * position.stock.inr_price,
            position.quantity * position.stock.inr_price / portfolio_value*100
        ])
        
    print(tabulate.tabulate(position_data,
             headers=["Ticker","Exchange","Quantity","Price","Market Value","% Allocation"],
             tablefmt="psql",
             floatfmt=".2f"
             ))
    
    print(f"total portfolio value: {portfolio_value:,.2f}.")
```
